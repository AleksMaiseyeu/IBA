```python
import random
import datetime
# import prettytable                  #   
import matplotlib.pyplot as plt     #   

def BubbleSort(A):                  #  
    for i in range(len(A)):
        for j in range(len(A)-1-i):
            if A[j] > A[j+1]:
                a = A[j]
                A[j] = A[j+1]
                A[j+1] = a


def QuickSort(A, fst, lst):         #  
    if fst >= lst:
        return

    #i, j = fst, lst
    pivot = A[fst]
    # pivot = A[random.randint(fst, lst)]
    first_bigger = fst+1
    while first_bigger <= lst:
        if A[first_bigger] >= pivot:
            break
        first_bigger += 1
    i = first_bigger+1
    while i <= lst:
        if A[i] < pivot:
            A[i], A[first_bigger] = A[first_bigger], A[i]
            first_bigger += 1
        i += 1

    last_smaller = first_bigger-1
    A[fst], A[last_smaller] = A[last_smaller], A[fst]
    QuickSort(A, fst, last_smaller-1)
    QuickSort(A, first_bigger, lst)

#  insert:
def InsertSort(A): 
        for i in range(len(A)):
          t = A[i]
          j = i
          while j >0 :
              if A[j-1]>t:
                 A[j]=A[j-1]
                 j-=1
              else:
                 break
              A[j]=t
        return A  

# ()  shaker   
def ShakerSort(A):
        for i in range(int(len(A) - 1 / 2)):
            for j in range(i, len(A) - 1 - i):
                if A[j] > A[j + 1]:
                    a = A[j]
                    A[j] = A[j + 1]
                    A[j + 1] = a

            for k in range(len(A) - 2 - i, -1):
                if A[k] <= A[k - 1]:
                    a = A[k]
                    A[k] = A[k - 1]
                    A[k - 1] = a
        return A



# table = prettytable.PrettyTable([" ", " ", " "," "," "])
x=[]
y1=[]
y2=[]
y3=[]
y4=[]


for N in range(1000,5001,1000):
    x.append(N)
    min=1
    max=N
    A=[]
    for i in range (N):
        A.append(int(round(random.random()*(max-min)+min)))

    #print(A)

    B = A.copy()
    # print(B)

    # BubbleSort(A)
    print("---")
    # print(A)



    # QuickSort(B, 0, len(B)-1)
    print("---")
    # print(B)

    t1 = datetime.datetime.now()
    BubbleSort(A)
    t2 = datetime.datetime.now()
    y1.append((t2-t1).total_seconds())
    print("    " +str(N)+"      "+str((t2-t1).total_seconds()) + "c")

    t3 = datetime.datetime.now()
    QuickSort(B, 0, len(B)-1)
    t4 = datetime.datetime.now()
    y2.append((t4 - t3).total_seconds())
    print("   " +str(N)+"      "+str((t4-t3).total_seconds()) + "c")


    t0 = datetime.datetime.now()
    InsertSort(B)
    t = datetime.datetime.now()
    y3.append((t - t0).total_seconds())
    print("   " +str(N)+"      "+str((t-t0).total_seconds()) + "c")

    
    t5 = datetime.datetime.now()
    ShakerSort(B)
    t6 = datetime.datetime.now()
    y4.append((t6 - t5).total_seconds())
    print("   " +str(N)+"      "+str((t6-t5).total_seconds()) + "c")


    # table.add_row([str(N), str((t2-t1).total_seconds()), str((t4-t3).total_seconds())])
# print(table)

font1 = {'family':'serif','color':'blue','size':12}
font2 = {'family':'serif','color':'darkred','size':10}

plt.title(" : ///", fontdict = font2)
plt.ylabel(", .", fontdict = font2)
plt.xlabel(" ", fontdict = font2)

plt.plot(x, y1, "C0", marker='o', markersize=7, label="")
plt.plot(x, y2, "C1", marker='o', markersize=7, label="")
plt.plot(x, y3, "C2", marker='o', markersize=7, label="")
plt.plot(x, y4, "C3", marker='o', markersize=7, label="")
#     
plt.legend(loc='best')
plt.show()


# 
#font1 = {'family':'serif','color':'blue','size':15}
#font2 = {'family':'serif','color':'red','size':12}

plt.title(" ", fontdict = font1)
plt.xlabel("", fontdict = font2)
plt.ylabel(" (.)", fontdict = font2)
plt.grid(True, linestyle='--', alpha=0.6)

plt.scatter(x, y1, color='C0')
plt.scatter(x, y2, color='C1')
plt.scatter(x, y3, color='C2')
plt.scatter(x, y4, color='C3')

plt.show()


# 
font1 = {'family':'serif','color':'blue','size':15}
font2 = {'family':'serif','color':'darkred','size':12}

plt.title(" ", fontdict = font1)
plt.xlabel(" ", fontdict = font2)
plt.ylabel(" (.)", fontdict = font2)
plt.grid(True, linestyle='--', alpha=0.6)

bar_width = 100
plt.bar([n - 1.5*bar_width for n in x], y1, width=bar_width, label='', color='C0')
plt.bar([n - 0.5*bar_width for n in x], y2, width=bar_width, label='', color='C1')
plt.bar([n + 0.5*bar_width for n in x], y3, width=bar_width, label='', color='C2')
plt.bar([n + 1.5*bar_width for n in x], y4, width=bar_width, label='', color='C3')

plt.show()

#   ---

plt.subplot(2, 2, 1)
plt.plot(x, y1, "C0")

plt.subplot(2, 2, 2)
plt.plot(x, y2, "C1")

plt.subplot(2, 2, 3)
plt.plot(x, y3, "C2")

plt.subplot(2, 2, 4)
plt.plot(x, y4, "C3")

plt.show()
```
